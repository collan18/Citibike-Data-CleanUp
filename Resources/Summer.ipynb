```python
all_filenames = [i for i in glob.glob(f"*{file_extension}")]
all_data = (pd.read_csv(f, sep=',') for f in all_filenames)
merged_data = pd.concat(all_data, ignore_index=True)
```

```python
combined_data = pd.DataFrame(merged_data)
```
